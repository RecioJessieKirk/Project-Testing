```python
import cv2
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt

#  Load the trained YOLOv5 model
model = YOLO("yolov5nu.pt")  # Ensure this is in the same folder as the notebook

#  Define the image path (change this to the image you want to test)
image_path = "pothole\images\potholes541.png"

#  Run inference
results = model(image_path)  # Detect objects in the image

#  Load the image
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct color display

#  Draw bounding boxes
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
        conf = box.conf[0]  # Confidence score
        cls = int(box.cls[0])  # Class index
        
        label = f"{model.names[cls]} {conf:.2f}"  # Class label + confidence
        color = (0, 255, 0)  # Green box color
        
        # Draw bounding box
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
        cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

#  Show the image with detections
plt.figure(figsize=(8, 6))
plt.imshow(image)
plt.axis("off")  # Hide axes
plt.title("Object Detection Result")
plt.show()
```
